## Ley de Planck

En términos de la frecuencia se expresa como:

$$ B_\nu(T) = \frac{2h\nu^3}{c^2} \frac{1}{e^{\frac{h\nu}{kT}} - 1} $$

Y en términos de la longitud de onda como:

$$ B_\lambda(T) = \frac{2hc^2}{\lambda^5} \frac{1}{e^{\frac{hc}{\lambda kT}} - 1} $$

```
#Importamos librerias
import numpy as np
from matplotlib import pyplot as plt
from astropy import constants as const

#Constantes
c = const.c.value    #Velocidad de la luz [m/s]
k = const.k_B.value  #Constante de Boltzmann [J/K]
h = const.h.value    #Constante de Planck [Js]

#Ley de Planck en términos de la frecuencia
def B_nu(nu, T):
  numerador   = 2 * h * nu**3
  denominador = c**2 * (np.exp(h * nu / (k * T)) - 1)
  return numerador/denominador

#Ley de Planck en términos de la longitud de onda
def B_lambda(lam, T):
    numerador   = 2 * h * c**2
    denominador = lam**5 * (np.exp((h * c) / (k * T * lam)) - 1)
    return numerador/denominador

#Graficamos la función de Planck en términos de la frecuencia
plt.plot(np.linspace(10e4,1e14, 1000), B_nu(np.linspace(10e4,1e14, 1000), 100)   , label="100 K")
plt.plot(np.linspace(10e4,1e16, 1000), B_nu(np.linspace(10e4,1e16, 1000), 1000)  , label="1000 K")
plt.plot(np.linspace(10e4,1e16, 1000), B_nu(np.linspace(10e4,1e16, 1000), 5000)  , label="5000 K")
plt.plot(np.linspace(10e4,1e17, 1000), B_nu(np.linspace(10e4,1e17, 1000), 10000) , label="10000 K")
plt.plot(np.linspace(10e4,1e17, 1000), B_nu(np.linspace(10e4,1e17, 1000), 50000) , label="50000 K")
plt.plot(np.linspace(10e4,1e18, 1000), B_nu(np.linspace(10e4,1e18, 1000), 100000), label="100000 K")

plt.xscale('log')
plt.yscale('log')

plt.title("Funcion de Plank en terminos de la frecuencia")
plt.xlabel(r"Frecuencia, $\nu$ (GHz)")
plt.ylabel(r"Brillo, B$_\nu$ (W m$^{-2}$ Hz$^{-1}$ Sr$^{-2}$)")

plt.ylim(1e-24)

plt.legend(frameon=False, loc='best')

plt.show()

#Graficamos la función de Planck en términos de la longitud de onda
plt.plot(np.logspace(-6, 3, 1000), B_lambda(np.logspace(-6, 3, 1000), 100)   , label="100 K")
plt.plot(np.logspace(-7, 3, 1000), B_lambda(np.logspace(-7, 3, 1000), 1000)  , label="1000 K")
plt.plot(np.logspace(-8, 3, 1000), B_lambda(np.logspace(-8, 3, 1000), 5000)  , label="5000 K")
plt.plot(np.logspace(-8, 3, 1000), B_lambda(np.logspace(-8, 3, 1000), 10000) , label="10000 K")
plt.plot(np.logspace(-9, 3, 1000), B_lambda(np.logspace(-9, 3, 1000), 50000) , label="50000 K")
plt.plot(np.logspace(-9, 3, 1000), B_lambda(np.logspace(-9, 3, 1000), 100000), label="100000 K")

plt.xscale('log')
plt.yscale('log')

plt.title("Funcion de Plank en terminos de la longitud de onda")
plt.xlabel(r"Longitud de onda, $\lambda$ (m)")
plt.ylabel(r"Brillo, B$_\nu$ (W m$^{-2}$ m$^{-1}$ Sr$^{-2}$)")

plt.ylim(1e-15)

plt.legend(frameon=False, loc='best')

plt.show()
```
